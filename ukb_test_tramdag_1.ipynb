# DATA

```python
import pandas as pd

df=pd.read_csv(DATA_PATH)
```

```python
df.head()
```

```python
pd.set_option('display.max_columns', None)
df.columns.tolist()
```

```python
df['Age_baseline | Instance 0'] = pd.to_datetime(df['Date of attending assessment centre | Instance 0']).dt.year - df['Year of birth']
df_sub = df[['Cholesterol | Instance 0',
             'Systolic blood pressure, automated reading | Instance 0 | Array 0',
             'Age_baseline | Instance 0']].dropna()
df_sub.info()

```

```python
df_sub = df_sub.rename(columns={
    'Cholesterol | Instance 0': 'chol',
    'Systolic blood pressure, automated reading | Instance 0 | Array 0': 'sys',
    'Age_baseline | Instance 0': 'age'
})
df_sample = df_sub.sample(n=10000, random_state=0)
```

```python
from sklearn.model_selection import train_test_split
train_df, temp_df = train_test_split(df_sample, test_size=0.2, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)
```

# API usage

## load configuration file and train model 

```python
from utils.tramdag import TramDagConfig , TramDagModel

cfg = TramDagConfig.load(CONF_DICT_PATH)
cfg.compute_levels(train_df)
device='cpu'
```

```python
# initalize model
td_model = TramDagModel.from_config(cfg, set_initial_weights=False,verbose=True,debug=True,device=device,initial_data = train_df) 
```

```python
td_model.fit(train_df, val_df,
            #  train_list=['target'],
             learning_rate=1e-3,
             epochs=3,batch_size=1_000,
             save_linear_shifts=True,
             save_simple_intercepts=True,
             verbose=False,debug=False,
             device=device,
             num_workers = 8,
             persistent_workers = True,
             prefetch_factor = 8,
             train_mode = "sequential")#,
```
