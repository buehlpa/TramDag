```python
import torch

def sample_standard_logistic(shape, epsilon=1e-7, device=None):
    uniform_samples = torch.rand(shape, device=device)
    clipped_uniform_samples = torch.clamp(uniform_samples, epsilon, 1 - epsilon)
    logistic_samples = torch.log(clipped_uniform_samples / (1 - clipped_uniform_samples))
    return logistic_samples
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import logistic

def sample_standard_logistic(shape, epsilon=1e-7, device=None):
    uniform_samples = torch.rand(shape, device=device)
    clipped = torch.clamp(uniform_samples, epsilon, 1 - epsilon)
    logistic_samples = torch.log(clipped / (1 - clipped))
    return logistic_samples

# Generate samples
num_samples = 10_000_00
samples = sample_standard_logistic((num_samples,)).cpu().numpy()

# Plot histogram of samples
plt.figure(figsize=(8, 5))
count, bins, _ = plt.hist(samples, bins=100, density=True, alpha=0.6, label='Sampled Histogram')

# Plot theoretical logistic PDF
x = np.linspace(-10, 10, 1000)
pdf = logistic.pdf(x)  # mean=0, scale=1 for standard logistic
plt.plot(x, pdf, 'r-', label='Theoretical PDF')

plt.title("Standard Logistic Distribution - Sampled vs Theoretical")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()





# Sample from the standard logistic distribution (mean=0, scale=1)
samples = logistic.rvs(size=100000)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(samples, bins=100, density=True, alpha=0.6, label='Sampled Histogram')

# Overlay the theoretical PDF
x = np.linspace(-10, 10, 1000)
pdf = logistic.pdf(x)
plt.plot(x, pdf, 'r-', label='Theoretical PDF')

plt.title("Standard Logistic Distribution - Scipy Samples vs PDF")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
